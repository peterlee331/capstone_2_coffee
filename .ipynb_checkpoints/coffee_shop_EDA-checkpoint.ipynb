```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
cof_data = pd.read_csv('post_dw_coffee_data.csv')
```

```python
cof_data.info()
```

```python
cof_data['Date'] = pd.to_datetime(cof_data['Date'])
```

```python
cof_data.info()
```

```python
cof_data.head()
```

```python
coffee_dat = cof_data.drop(columns=['Unnamed: 0'], inplace=False)
```

```python
coffee_dat.isnull().sum()
```

```python
coffee_dat.describe().T
```

```python
coffee_dat.shape
```

```python
coffee_dat['money'].hist()
plt.title('Distribution of Revenue per Transaction')
plt.show()
```

```python
coffee_dat['money'].skew()
```

```python
coffee_dat.boxplot(column='money')
plt.show()
```

```python
coffee_dat['hour_of_day'].value_counts().plot(kind='bar')
plt.title('hour_of_day Frequency')
plt.show()
```

```python
coffee_dat['coffee_name'].value_counts().plot(kind='bar')
plt.title('coffee_name Frequency')
plt.show()
```

```python
coffee_dat['Time_of_Day'].value_counts().plot(kind='bar')
plt.title('Time_of_Day Frequency')
plt.show()
```

```python
coffee_dat['Weekday'].value_counts().plot(kind='bar')
plt.title('Weekday Frequency')
plt.show()
```

```python
coffee_dat['Month_name'].value_counts().plot(kind='bar')
plt.title('Month Frequency')
plt.show()
```

## Sales by Month

```python
coffee_dat.groupby('Monthsort')['money'].sum()
```

```python
coffee_dat.groupby('Monthsort')['money'].mean()
```

## Sales by Weekday

```python
coffee_dat.groupby('Weekday')['money'].sum().sort_values(ascending=False)
```

## Sales by Hour_of_day

```python
coffee_dat.groupby('hour_of_day')['money'].sum().sort_values(ascending=False)
```

```python
daily_sales = coffee_dat.groupby('Date')['money'].sum()
rolling_sales = daily_sales.rolling(7).mean()
plt.figure(figsize=(14,6))
plt.plot(daily_sales)
plt.plot(rolling_sales)
plt.title('Daily Coffee Sales')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(coffee_dat['hour_of_day'], coffee_dat['money'])
plt.xlabel('Hour of Day')
plt.ylabel('Money')
plt.title('Revenue vs Hour of Day')
plt.show()
```

## Mean Revenue by Weekday

```python
coffee_dat.groupby('Weekday')['money'].mean().sort_values()
```

```python
grouped = coffee_dat.groupby(['hour_of_day', 'Weekday'])['money'].mean().unstack()

plt.figure(figsize=(12,6))
grouped.plot()
plt.title('Interaction: Hour of Day  Weekday')
plt.ylabel('Average Revenue')
plt.show()
```

```python
plt.figure(figsize=(9,7))
sns.heatmap(coffee_dat.corr(numeric_only=True),linewidths=.1,cmap="YlGnBu", annot=True)
plt.yticks(rotation=0);
```

```python
coffee_dat.boxplot(column='money', by='coffee_name', rot=90)
plt.show()
```
